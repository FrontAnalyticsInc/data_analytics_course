## Create a script to parse this csv file and create a new csv file for each table

```python
with open('/Users/dpl/Downloads/Section1all_csv.csv', 'r') as f:
    c = 0
    num_files = 0
    in_table = False
    for line in f:
        c += 1
        # print(len(line.split(',')))
        row_list = line.split(',')
        if not in_table:
            if row_list[0] == 'Line':
                in_table = True
                new_file = open('output_file_{0:04d}.csv'.format(num_files), 'w')
        else:
            if row_list[0] != '':
                new_file.write(line)
            else:
                new_file.close()
                num_files += 1
                in_table = False
print(num_files)
```

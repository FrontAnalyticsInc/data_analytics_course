                    <div style="text-indent: -1em; padding-left: 1em;">The ivy across our back fence tangles gray <br></div><div style="text-indent: -1em; padding-left: 1em;">into a green evening light.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">How a second emptiness <br></div><div style="text-indent: -1em; padding-left: 1em;">un-punctuates the first.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">Disloyal,<br></div><div style="text-indent: -1em; padding-left: 1em;">we attempt to construct.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">An ache will tighten <br></div><div style="text-indent: -1em; padding-left: 1em;">but not form.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">Making impossible<br></div><div style="text-indent: -1em; padding-left: 1em;">even this upsurge of crows across our sightline.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">The Mayans invented zero so as not to ignore even the gods <br></div><div style="text-indent: -1em; padding-left: 1em;">who <em>wouldn't </em>carry their burdens.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">Too slippery as prayer, too effortless <br></div><div style="text-indent: -1em; padding-left: 1em;">as longing.<br></div><div style="text-indent: -1em; padding-left: 1em;"><br></div><div style="text-indent: -1em; padding-left: 1em;">Our problem was preparation.                                                                     <div class="feature-bd">
                                        <p>Poet, critic, and publisher Rusty Morrison was a teacher for 19 years before earning her MFA from Saint Marys College of California. Her first book, <em>Whethering </em>(2004), won the Colorado Prize for Poetry; her second, <em>the true keeps calm biding its story </em>(2008), won a Sawtooth Poetry Prize, a James Laughlin Award, a Northern California Book Award, and a Di Castagnola Award from the Poetry Society of America. Morrisons other works include <em>After Urgency </em>(2012), winner of the Dorset Prize, <em>Book of the Given </em>(2012), and <em>Beyond the Chainlink </em>(2014). Known for her innovative use and re-use of form and line, Morrison has been praised for, in the words of Brian Teare, the way she pairs lyric's mimesis of interiority with philosophy's relentless self-scrutiny. Her work has been anthologized in the <em>Postmodern American Poetry: A Norton Anthology </em>(2nd Edition, 2013), <em>The Arcadia Project: North American Postmodern Pastoral </em>(2012), <em>The Sonnets:...</em></p>

```python
def create_unique_id(df, colname):
    id_column_name = colname.lower() + '_id'
    lookup_table = pd.DataFrame(list(
                      zip(df[colname].unique(), 
                         [str(uuid.uuid1()) for x in range(len(df[colname].unique()))]
                               )), 
                    columns=[colname, id_column_name]).set_index(colname)
    print(lookup_table)
    df[id_column_name] = df[colname].apply(lambda x: lookup_table.loc[x])
create_unique_id(df, 'Supplies Subgroup')
df.head()
```
